```python
import requests
import json
import time
```

```python
nyt_key = json.load(open("api_key.json"))['nyt']
```

```python
authors = open("authors.txt").read().split('\n')
# get rid of empty last author
authors = authors[:-1]
authors
```

```python
def nytAPICall(name):
    url = "https://api.nytimes.com/svc/books/v3/reviews.json"
    payload = {'api-key': nyt_key, 'author' : name}
    r = requests.get(url, params=payload)
    return r.json()
```

```python
authors_json = []
for author in authors:
    print(author)
    author_data = {'name' : author, 'nyt' : nytAPICall(author)['results']}
    authors_json.append(author_data)
    time.sleep(12)
```

```python
authors_json
```

```python
with open('author_data.json', 'w') as outfile:
    json.dump(authors_json, outfile)
```
